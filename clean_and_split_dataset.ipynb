# 划分数据集

```python
training_set_files, validation_set_files = du.split_dataset_into_training_and_test_sets(
    data_set_dir='datasets/all_train', 
    training_set_dir='datasets/train',
    test_set_dir='datasets/validation',
    test_set_size=0.1)
```

```python
def gcd(a, b):
    """ Calculate the GCD of a and b
    """
    if b == 0:
        return a
    else:
        return gcd(b, a % b)

print('The GCD of the training set files and validation set files: {}'.format(
    gcd(training_set_files, validation_set_files)))
```
