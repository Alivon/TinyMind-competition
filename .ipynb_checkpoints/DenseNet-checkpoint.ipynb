```python
from recognition import train_model as tm
from recognition import evaluate_model as em
from recognition import csv_utils as cu
from recognition import model_utils as mu


%matplotlib inline
```

```python
# global variables
BATCH_SIZE = 16 # batch size
IMAGE_SIZE = 224 # image size
COLOR_MODE = 'grayscale' # color mode 
RESCALE = 1./255
```

```python
train_generator = mu.my_generator('train', 
                                  rescale=RESCALE,
                                  horizontal_flip=False,
                                  vertical_flip=False,
                                  width_shift_range=0.1,
                                  height_shift_range=0.1,
                                  rotation_range=10,
                                  shear_range=10,
                                  zoom_range=0.1,
                                  shuffle=True,
                                  batch_size=BATCH_SIZE, 
                                  color_mode=COLOR_MODE, 
                                  target_size=(IMAGE_SIZE, IMAGE_SIZE))

valid_generator = mu.my_generator('datasets/validation', 
                                  rescale=RESCALE,
                                  shuffle=True,
                                  batch_size=BATCH_SIZE, 
                                  color_mode=COLOR_MODE, 
                                  target_size=(IMAGE_SIZE, IMAGE_SIZE))

print('Writing the class_indices into csv file.')
cu.write_into_csv(train_generator.class_indices, csv_path='class_indices.csv')
print('OK!')
```

# DenseNet

```python
from keras.applications.densenet import DenseNet201, DenseNet169, DenseNet121
from keras.layers import BatchNormalization, Activation, Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, PReLU, GlobalAveragePooling2D
from keras.models import Model
from keras.regularizers import l2


def build_model_based_on_densenet(out_dims, 
                                  base_model=DenseNet201, 
                                  base_weights=None, 
                                  input_shape=(224, 224, 3), 
                                  weights=None):
    """
    Inputs:
      - out_dims:
      
    Return:
       - model: 
    """
    densenet_base_model = base_model(weights=base_weights, 
                                     include_top=False, 
                                     input_shape=input_shape)
    x = densenet_base_model.output
    
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = PReLU()(x)
    x = Dropout(0.4)(x)
    
    x = Dense(512)(x)
    x = BatchNormalization()(x)
    x = PReLU()(x)
    x = Dropout(0.4)(x)
    
    x = Dense(out_dims)(x)
    x = Activation('softmax')(x)
    model = Model(inputs=densenet_base_model.input, outputs=x)
    
    if weights:
        model.load_weights(weights, by_name=True)

    return model

mu.clear_session()

channels = mu.get_channels(COLOR_MODE)
if channels is 1:
    base_weights = None
else:
    base_weights = 'imagenet'
    
model = build_model_based_on_densenet(100, 
                                      base_model=DenseNet121,
                                      base_weights=base_weights, 
                                      input_shape=(IMAGE_SIZE, IMAGE_SIZE, channels), 
                                      weights='models/DenseNet/121/model4.h5')
for index, layer in enumerate(model.layers):
    print(index, layer.name)
    
print('\n The architecture of the model:')
model.summary()
```

```python
freeze_range = range(0, 0)
model = mu.freeze_layers(model, freeze_range)
mu.count_parameters(model) 
```

```python
LEARNING_RATE = 0.00001
DECAY = 0
EPOCHS = 8

model_path = 'models/DenseNet/121/model6.h5'
history = tm.train_model(model, 
                         train_generator, 
                         valid_generator, 
                         model_path=model_path, 
                         patience=1,
                         reduce_factor=0.1,
                         reduce_time=8,
                         reduce_monitor='loss',
                         save_best='loss',
                         batch_size=BATCH_SIZE, 
                         learning_rate=LEARNING_RATE, 
                         decay=DECAY, 
                         epochs=EPOCHS)
```

# Draw history curve

```python
mu.draw_plot(history)
```
