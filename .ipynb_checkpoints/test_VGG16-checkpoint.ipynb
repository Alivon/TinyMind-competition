# Data augmentation at testing stage.

```python
generator = solver.my_generator('dataset/czy',
                                shuffle=False,
                                rescale=RESCALE, 
                                batch_size=BATCH_SIZE, 
                                target_size=(IMAGE_SIZE, IMAGE_SIZE), 
                                color_mode=COLOR_MODE)

prediction = solver.get_prediction_by_single_generator(model, generator)
    
class_indices = solver.get_class_indices('class_indices.csv')
for k in [1, 5]:
    topk_classes = solver.get_topk_classes_by_single_generator(generator, class_indices, prediction=prediction, k=k)
    total_number, correct_number = 0, 0
    for topk_class, filename in zip(topk_classes, generator.filenames):
        if filename[:1] in topk_class:
            correct_number += 1
        total_number += 1
    print('Top-{} accuracy: {:.2%}'.format(k, correct_number / total_number))
```

```python
import cv2
import os
import numpy as np


def cv_imread(image_path, color_mode='rgb'):
    """ uses opencv to read an image which path contains chinese
    Inputs:
      - image_path: A string, path of image
      - color_mode: 'rgb' or 'grayscale'
      
    Return:
      - image_data: A numpy array, data of image
    """
    if color_mode not in {'rgb', 'grayscale'}:
        raise ValueError('color_mode must be \'rgb\' or \'grayscale\'.')
        
    if color_mode == 'rgb':
        image_data = cv2.imdecode(np.fromfile(image_path, dtype=np.uint8), cv2.IMREAD_COLOR)
    else:
        image_data = cv2.imdecode(np.fromfile(image_path, dtype=np.uint8), cv2.IMREAD_GRAYSCALE)
    return image_data


def rotate_image_by_angle(image, 
                          need_resize=False, 
                          target_size=(224, 224), 
                          fill_color=(255, 255, 255),
                          angle=0):
    """
    Inputs: 
      - image: A numpy array contains image data with 
              shape(height, width, channels).
      - angle: A float number between 0 and 360.
      
    Return:
      - rotate_image: 
    """
    height, width = image.shape[0], image.shape[1]
    
    center = (height//2, width//2)
    matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotate_image = cv2.warpAffine(image, 
                                  matrix, 
                                  (width, height), 
                                  borderValue=fill_color)
    if need_resize:
        rotate_image = cv2.resize(rotate_image, target_size)
    return rotate_image


def predict_by_augment_patches(model,
                               data_path,
                               generator, 
                               show_first_fig=False,
                               rescale=1./255,
                               color_mode='rgb', 
                               target_size=(224, 224)):
    """
    Inputs:
      - model:
      - data_path:
      - generator:
      - show_first_fig:
      - ratio:
      - rescale:
      - color_mode:
      - target_size:
      
    Return:
      - prediction:
    """
    prediction = []
    for file in generator.filenames:
        filepath = os.path.join(data_path, file)

        if color_mode == 'rgb':
            image_data = cv_imread(filepath, color_mode=color_mode)
            height, width, _ = image_data.shape
        else:
            image_data = cv_imread(filepath, color_mode=color_mode)
            height, width = image_data.shape

        augment_patches = []
        weights = np.array([[8./24, 
                             1./24, 1./24, 1./24, 1./24, 1./24, 
                             1./24, 1./24, 1./24, 1./24, 1./24, 
                             ]]).T
        # original
        augment_patches.append(cv2.resize(image_data, target_size))
        
        # 0.8 crop
        crop_height = int(height*0.8)
        crop_width = int(width*0.8)
        margin_height = (height-crop_height) // 2
        margin_width = (width-crop_width) // 2
        middle = image_data[margin_height:height-margin_height,
                           margin_width:width-margin_width]
        middle = cv2.resize(middle, target_size)
        augment_patches.append(middle)
        
        left = image_data[:, :crop_width]
        left = cv2.resize(left, target_size)
        augment_patches.append(left)

        right = image_data[:, width-crop_width:]
        right = cv2.resize(right, target_size)
        augment_patches.append(right)
        
        top = image_data[:crop_height, :]
        top = cv2.resize(top, target_size)
        augment_patches.append(top)
        
        bottom = image_data[height-crop_height:, :]
        bottom = cv2.resize(bottom, target_size)
        augment_patches.append(bottom)
        
        # 0.6 crop
         # middle
        crop_height = int(height*0.6)
        crop_width = int(width*0.6)
        margin_height = (height-crop_height) // 2
        margin_width = (width-crop_width) // 2
        middle = image_data[margin_height:height-margin_height,
                           margin_width:width-margin_width]
        middle = cv2.resize(middle, target_size)
        augment_patches.append(middle)
        
        left = image_data[:, :crop_width]
        left = cv2.resize(left, target_size)
        augment_patches.append(left)

        right = image_data[:, width-crop_width:]
        right = cv2.resize(right, target_size)
        augment_patches.append(right)
        
        top = image_data[:crop_height, :]
        top = cv2.resize(top, target_size)
        augment_patches.append(top)
        
        bottom = image_data[height-crop_height:, :]
        bottom = cv2.resize(bottom, target_size)
        augment_patches.append(bottom)
        
        # Top-1 accuracy: 95.56%
        # Top-5 accuracy: 98.61%
        '''
        left_top = image_data[:crop_height, :crop_width]
        left_top = cv2.resize(left_top, target_size)
        augment_patches.append(left_top)

        left_bottom = image_data[height-crop_height:, :crop_width]
        left_bottom = cv2.resize(left_bottom, target_size)
        augment_patches.append(left_bottom)

        right_top = image_data[:crop_height, width-crop_width:]
        right_top = cv2.resize(right_top, target_size)
        augment_patches.append(right_top)

        right_bottom = image_data[height-crop_height:, width-crop_width:]
        right_bottom = cv2.resize(right_bottom, target_size)
        augment_patches.append(right_bottom)
        # Top-1 accuracy: 95.51%
        # Top-5 accuracy: 98.56%
        '''
        
        '''
        # rotation 10
        rotate = rotate_image_by_angle(image_data, 
                                       need_resize=True, 
                                       target_size=target_size, 
                                       angle=10,
                                       fill_color=(255, 255, 255))
        augment_patches.append(rotate)
        # rotation -10
        rotate = rotate_image_by_angle(image_data, 
                                       need_resize=True, 
                                       target_size=target_size, 
                                       angle=-10,
                                       fill_color=(255, 255, 255))
        augment_patches.append(rotate)
        '''
        
        if show_first_fig is True:
            patch_size = len(augment_patches)
            plt.figure(figsize=(4*patch_size, 4*patch_size))
            for index, patch in enumerate(augment_patches):
                plt.subplot(1, patch_size, index + 1)
                plt.imshow(patch)
                plt.axis('off')
        
            break
            
        if color_mode == 'grayscale':
            augment_patches = np.expand_dims(augment_patches, axis=3)
        
        augment_patches = np.array(augment_patches)
        augment_patches = augment_patches * rescale
        
        mean_prediction = np.mean(model.predict(augment_patches) * weights, 
                                  axis=0)
        prediction.append(mean_prediction)
        
    return prediction
```

```python
prediction = predict_by_augment_patches(model,
                                        'dataset/czy',
                                        generator, 
                                        show_first_fig=True,
                                        rescale=RESCALE,
                                        color_mode=COLOR_MODE, 
                                        target_size=(IMAGE_SIZE, IMAGE_SIZE))
```

```python
prediction = predict_by_augment_patches(model,
                                        'dataset/czy',
                                        generator, 
                                        show_first_fig=False,
                                        rescale=RESCALE,
                                        color_mode=COLOR_MODE, 
                                        target_size=(IMAGE_SIZE, IMAGE_SIZE))

print('Validation accuracy after data augmentation at test stage:')

for k in [1, 5]:
    topk_classes = solver.get_topk_classes_by_single_generator(generator, class_indices, prediction=prediction, k=k)
    total_number, correct_number = 0, 0
    for topk_class, filename in zip(topk_classes, generator.filenames):
        if filename[:1] in topk_class:
            correct_number += 1
        total_number += 1
    print('Top-{} accuracy: {:.2%}'.format(k, correct_number / total_number))
```
