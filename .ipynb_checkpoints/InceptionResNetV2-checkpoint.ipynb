```python
from recognition import model_utils as mu
from recognition import train_model as tm
from recognition import evaluate_model as em
from recognition import csv_utils as cu


%matplotlib inline
```

```python
# global variables
BATCH_SIZE = 64 # batch size
IMAGE_SIZE = 160 # image size
COLOR_MODE = 'grayscale' # color mode 
RESCALE = 1./255
```

```python
train_generator = mu.my_generator('datasets/all_train', 
                                  rescale=RESCALE,
                                  horizontal_flip=False,
                                  vertical_flip=False,
                                  rotation_range=10,
                                  width_shift_range=0.1,
                                  height_shift_range=0.1,
                                  shear_range=10.0,
                                  zoom_range=0.1,
                                  shuffle=True,
                                  channel_shift_range=0.0,
                                  batch_size=BATCH_SIZE, 
                                  color_mode=COLOR_MODE, 
                                  target_size=(IMAGE_SIZE, IMAGE_SIZE))

valid_generator = mu.my_generator('datasets/validation', 
                                  rescale=RESCALE,
                                  shuffle=True,
                                  batch_size=BATCH_SIZE, 
                                  color_mode=COLOR_MODE, 
                                  target_size=(IMAGE_SIZE, IMAGE_SIZE))

print('Writing the class_indices into csv file.')
cu.write_list_or_dict_into_csv(train_generator.class_indices, csv_path='class_indices.csv', have_chinese=False)
print('OK!')
```

# InceptionResNetV2

```python
from keras.applications.inception_resnet_v2 import InceptionResNetV2
from keras.layers import BatchNormalization, Activation, Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, PReLU, GlobalAveragePooling2D
from keras.models import Model
from keras.regularizers import l2


def build_model_based_on_inception_resnet_v2(out_dims, 
                                             base_weights=None, 
                                             input_shape=(299, 299, 3), 
                                             weights=None):
    """
    Input:
      - out_dims:
      
    Return:
       - model: 
    """
    inception_resnet_v2_base_model = InceptionResNetV2(weights=base_weights, 
                                                       include_top=False, 
                                                       input_shape=input_shape)
    x = inception_resnet_v2_base_model.output
    
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = PReLU()(x)
    x = Dropout(0.6)(x)
    
    x = Dense(512)(x)
    x = BatchNormalization()(x)
    x = PReLU()(x)
    x = Dropout(0.6)(x)
    
    x = Dense(out_dims)(x)
    x = Activation('softmax')(x)
    model = Model(inputs=inception_resnet_v2_base_model.input, outputs=x)
    
    if weights:
        model.load_weights(weights, by_name=False)

    return model

mu.clear_session()

channels = mu.get_channels(COLOR_MODE)
if channels is 1:
    base_weights = None
else:
    base_weights = 'imagenet'
model = build_model_based_on_inception_resnet_v2(100, 
                                                 base_weights=base_weights, 
                                                 input_shape=(IMAGE_SIZE, IMAGE_SIZE, channels), 
                                                 weights='models/InceptionResNetV2/model3.h5')

for index, layer in enumerate(model.layers):
    print(index, layer.name)
    
print('\n The architecture of the model:')
model.summary()
```

```python
freeze_range = range(0, 0)
model = mu.freeze_layers(model, freeze_range) # 780
mu.count_parameters(model) 
```

```python
LEARNING_RATE = 0.00001
DECAY = 0
EPOCHS = 8

model_path = 'models/InceptionResNetV2/model5.h5'
history = tm.train_model(model,
                         train_generator, 
                         valid_generator,
                         model_path=model_path,
                         print_lr=False,
                         automatic_reduction=True,
                         patience=3,
                         reduce_factor=0.1,
                         reduce_time=2,
                         reduce_monitor='val_acc',
                         save_best='val_loss',
                         batch_size=BATCH_SIZE,
                         learning_rate=LEARNING_RATE,
                         decay=DECAY,
                         epochs=EPOCHS)
```

# Draw history curve

```python
mu.draw_plot(history)
```

# Evaluate the performance of our model

### Load the model

```python
model = mu.load_my_model('models/InceptionResNetV2/model3.h5')
```

```python
model.save('models/InceptionResNetV2/model6.h5')
```

### Get the mapping from classes to indices

```python
class_indices = cu.get_class_indices('class_indices.csv')
print(class_indices)
```

### Get the prediction for the validation set

```python
generator = mu.my_generator('datasets/validation',
                             shuffle=False,
                             rescale=RESCALE, 
                             batch_size=BATCH_SIZE, 
                             target_size=(IMAGE_SIZE, IMAGE_SIZE), 
                             color_mode=COLOR_MODE)


use_augment = False
if use_augment:
    prediction = em.predict_by_augment_patches(model,
                                               'datasets/validation',
                                               generator, 
                                               batch_size=32, 
                                               num_classes=100,
                                               show_first_fig=False,
                                               rescale=RESCALE,
                                               color_mode=COLOR_MODE, 
                                               target_size=(IMAGE_SIZE, IMAGE_SIZE))
else:
    prediction = em.get_prediction_by_single_generator(model, generator)
print('OK!')
```

### Evaluate the model

```python
output = {}
for k in [1, 5]:
    output[str(k)] = em.evaluate_topk_accuracy_by_single_generator(generator, 
                                                                   class_indices,
                                                                   prediction=prediction, 
                                                                   k=k)
```
