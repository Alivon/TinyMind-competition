```python
from recognition import model_utils as mu
from recognition import train_model as tm
from recognition import evaluate_model as em
from recognition import csv_utils as cu


%matplotlib inline
```

# Obtain the data generator

```python
# global variables
BATCH_SIZE = 32 # batch size
IMAGE_SIZE = 180 # image size
COLOR_MODE = 'grayscale' # color mode 
RESCALE = 1./255
```

```python
train_generator = mu.my_generator('datasets/all_train', 
                                  rescale=RESCALE,
                                  horizontal_flip=False,
                                  vertical_flip=False,
                                  rotation_range=10,
                                  width_shift_range=0.1,
                                  height_shift_range=0.1,
                                  # shear_range=10.0,
                                  zoom_range=0.1,
                                  shuffle=True,
                                  channel_shift_range=0.0,
                                  batch_size=BATCH_SIZE, 
                                  color_mode=COLOR_MODE, 
                                  target_size=(IMAGE_SIZE, IMAGE_SIZE))

valid_generator = mu.my_generator('datasets/validation', 
                                  rescale=RESCALE,
                                  shuffle=True,
                                  batch_size=BATCH_SIZE, 
                                  color_mode=COLOR_MODE, 
                                  target_size=(IMAGE_SIZE, IMAGE_SIZE))

print('Writing the class_indices into csv file.')
cu.write_list_or_dict_into_csv(train_generator.class_indices, csv_path='class_indices.csv', have_chinese=False)
print('OK!')
```

# Build our model

```python
from keras.applications.xception import Xception
from keras.layers import GlobalAveragePooling2D, Dense, Activation, Flatten, Conv2D, Dropout, BatchNormalization, PReLU
from keras.models import Model
from keras.regularizers import l2


def build_model_based_on_xception(out_dims, 
                                  base_weights=None, 
                                  input_shape=(229, 229, 3), 
                                  weights=None):
    
    base_xception_model = Xception(weights=base_weights, 
                                   include_top=False, 
                                   input_shape=input_shape)
    x = base_xception_model.output
    
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = PReLU()(x)
    # x = Dropout(0.4)(x)
    
    x = Dense(512)(x)
    x = BatchNormalization()(x)
    x = PReLU()(x)
    x = Dropout(0.4)(x)
    
    x = Dense(out_dims, kernel_regularizer=l2(0.01))(x)
    x = Activation('softmax')(x)
    
    model = Model(inputs=base_xception_model.input, outputs=x)
    
    if weights:
        model.load_weights(weights, by_name=False)
        
    return model


mu.clear_session()
channels = mu.get_channels(COLOR_MODE)

if channels is 1:
    base_weights = None
else:
    base_weights = 'imagenet'

model = build_model_based_on_xception(100, 
                                      base_weights=base_weights, 
                                      input_shape=(IMAGE_SIZE, IMAGE_SIZE, channels), 
                                      weights='models/Xception/model17.h5')

for index, layer in enumerate(model.layers):
    print(index, layer.name)
    
print('\nThe architecture of the model:')
model.summary()
```

# Freeze some layers and check the parameters again

```python
freeze_range = range(0, 0)
model = mu.freeze_layers(model, freeze_range)
mu.count_parameters(model) 
```

# Select parameters for training

```python
LEARNING_RATE = 0.000001
DECAY = 0
EPOCHS = 8

model_path = 'models/Xception/model21.h5' 
history = tm.train_model(model,
                         train_generator, 
                         valid_generator,
                         model_path=model_path,
                         print_lr=False,
                         automatic_reduction=True,
                         patience=3,
                         reduce_factor=0.1,
                         reduce_time=2,
                         reduce_monitor='val_acc',
                         save_best='val_loss',
                         batch_size=BATCH_SIZE,
                         learning_rate=LEARNING_RATE,
                         decay=DECAY,
                         epochs=EPOCHS)
```

# Draw history curve

```python
mu.draw_plot(history)
```

# Evaluate the performance of our model.

### Load model

```python
model = mu.load_my_model('models/Xception/model17.h5')
model.summary()
```

### Get the mapping from classes to indices

```python
class_indices = cu.get_class_indices('class_indices.csv')
print(class_indices)
```

### Get the prediction for the validation set

```python
generator = mu.my_generator('datasets/validation',
                             shuffle=False,
                             rescale=RESCALE, 
                             batch_size=BATCH_SIZE, 
                             target_size=(IMAGE_SIZE, IMAGE_SIZE), 
                             color_mode=COLOR_MODE)


use_augment = False
if use_augment:
    prediction = em.predict_by_augment_patches(model,
                                               'datasets/validation',
                                               generator, 
                                               batch_size=32, 
                                               num_classes=100,
                                               show_first_fig=False,
                                               rescale=RESCALE,
                                               color_mode=COLOR_MODE, 
                                               target_size=(IMAGE_SIZE, IMAGE_SIZE))
else:
    prediction = em.get_prediction_by_single_generator(model, generator)
print('OK!')
```

### Evaluate the model

```python
output = {}
for k in [1, 5]:
    output[str(k)] = em.evaluate_topk_accuracy_by_single_generator(generator, 
                                                                   class_indices,
                                                                   prediction=prediction, 
                                                                   k=k)
```

### Check the wrong information

```python
correct_number, total_number, topk_indices, topk_classes, wrong_info = output[str(1)]
for info in wrong_info:
    print(info)
```

### Visualize the wrong images

```python
from recognition import image_utils as iu

iu.show_images_in_wrong_info(wrong_info, 'train', number=20)
```

# Write answer in the csv file

```python
test_generator = mu.my_generator('datasets/test2',
                                 shuffle=False,
                                 rescale=RESCALE, 
                                 batch_size=BATCH_SIZE, 
                                 target_size=(IMAGE_SIZE, IMAGE_SIZE), 
                                 color_mode=COLOR_MODE)

use_augment = True

if use_augment:
    prediction = em.predict_by_augment_patches(model,
                                               'datasets/test2',
                                               test_generator, 
                                               batch_size=32, 
                                               num_classes=100,
                                               show_first_fig=False,
                                               rescale=RESCALE,
                                               color_mode=COLOR_MODE, 
                                               target_size=(IMAGE_SIZE, IMAGE_SIZE))
else:
    prediction = em.get_prediction_by_single_generator(model, 
                                                       test_generator)
    
topk_indices = em.get_topk_indices_by_single_generator(test_generator, 
                                                       prediction=prediction, 
                                                       k=5)

results = cu.get_pd_data(test_generator, 
                         class_indices, 
                         topk_indices=topk_indices)
cu.write_pd_data_into_csv(results, 
                          'results/Xception/results_Xception.csv', 
                          have_chinese=True, 
                          index=None, 
                          header=True)
print('OK!')
```
